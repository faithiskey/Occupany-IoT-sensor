```python
# Convert Date to datetime64 and then to float (timestamp)
feature['date'] = pd.to_datetime(feature['date'])
feature['date'] = feature['date'].apply(lambda x: x.timestamp())

# Split data into training and testing sets
from sklearn.model_selection import train_test_split
feature_train, feature_test, Target_train, Target_test = train_test_split(feature[['date', 'Light']], Target, test_size=0.2, random_state=42)
```

```python
# Predict on test set
Target_pred = model.predict(feature_test)
```

```python
# Evaluate accuracy
accuracy = accuracy_score(Target_test, Target_pred)
print("Accuracy:", accuracy)
```
